```python
merged_df[["강수량(mm)", "3시간신적설(cm)"]] = merged_df[["강수량(mm)", "3시간신적설(cm)"]].fillna(0)
merged_df["avg_temp"] = merged_df[["5cm 지중온도(°C)",	"10cm 지중온도(°C)",	"20cm 지중온도(°C)",	"30cm 지중온도(°C)"]].mean(axis=1)
merged_df = merged_df.drop(["마명","5cm 지중온도(°C)",	"10cm 지중온도(°C)",	"20cm 지중온도(°C)",	"30cm 지중온도(°C)"], axis=1)
merged_df
```

```python
merged_df["등급"] = merged_df["등급"].str.extract('(\d+)')
merged_df.head(20)
```

```python
merged_df["마필성별"] = merged_df["마필성별"].replace({'수': -1, '암': 1, '거': 0})
merged_df["혈통"] = merged_df["혈통"].replace({'제주': 1, '한라': 0})
merged_df["체중증감"] = merged_df["체중증감"].fillna(0)
merged_df
```
