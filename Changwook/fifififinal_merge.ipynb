```python
train_data=data[data['경주일자']<=20220631]
test_data=data[data['경주일자']>=20220701]
```

```python
train_data_1200=train_data[train_data['경주거리']==1200]
test_data_1200=test_data[test_data['경주거리']==1200]
train_data_1200

#기온, 지면온도, avg_temp중 하나
#
```

```python
train_data_1200.loc[train_data_1200['주로상태_함수율']<=4,'주로상태_함수율']=0
train_data_1200.loc[(train_data_1200['주로상태_함수율']>4)&(train_data_1200['주로상태_함수율']<8),'주로상태_함수율']=1
train_data_1200.loc[(train_data_1200['주로상태_함수율']>=8)&(train_data_1200['주로상태_함수율']<10),'주로상태_함수율']=2
train_data_1200.loc[(train_data_1200['주로상태_함수율']>=10),'주로상태_함수율']=3


test_data_1200.loc[test_data_1200['주로상태_함수율']<=4,'주로상태_함수율']=0
test_data_1200.loc[(test_data_1200['주로상태_함수율']>4)&(test_data_1200['주로상태_함수율']<8),'주로상태_함수율']=1
test_data_1200.loc[(test_data_1200['주로상태_함수율']>=8)&(test_data_1200['주로상태_함수율']<10),'주로상태_함수율']=2
test_data_1200.loc[(test_data_1200['주로상태_함수율']>=10),'주로상태_함수율']=3
```

```python
train_data_1200["key"]=0
unique = train_data_1200[["경주일자",	"경주번호"]].drop_duplicates()
unique = unique.reset_index()
unique

for i in range(len(unique)):
    train_data_1200.loc[(train_data_1200["경주일자"]==unique["경주일자"][i])&(train_data_1200["경주번호"]==unique["경주번호"][i]), "key"] = i

test_data_1200["key"]=0
unique = test_data_1200[["경주일자",	"경주번호"]].drop_duplicates()
unique = unique.reset_index()
unique

for i in range(len(unique)):
    test_data_1200.loc[(test_data_1200["경주일자"]==unique["경주일자"][i])&(test_data_1200["경주번호"]==unique["경주번호"][i]), "key"] = i
```

```python
tot_mean_1200=np.mean(train_data_1200[train_data_1200['착순']==1]['경주기록(초)'])

keys=np.unique(train_data_1200['key'])
train_data_1200['adj_record']=0
for key in keys:
    mask_key = (train_data_1200['key'] == key)
    mask_win = (train_data_1200['key'] == key) & (train_data_1200['착순'] == 1)
    
    train_data_1200.loc[mask_key, 'adj_record'] = (
        train_data_1200.loc[mask_key, '경주기록(초)'].values *
        tot_mean_1200/train_data_1200.loc[mask_win, '경주기록(초)'].values[0]
    )

keys=np.unique(test_data_1200['key'])
test_data_1200['adj_record']=0
for key in keys:
    mask_key = (test_data_1200['key'] == key)
    mask_win = (test_data_1200['key'] == key) & (test_data_1200['착순'] == 1)
    
    test_data_1200.loc[mask_key, 'adj_record'] = (
        test_data_1200.loc[mask_key, '경주기록(초)'].values *
        tot_mean_1200/test_data_1200.loc[mask_win, '경주기록(초)'].values[0]
    )
```

```python
road=np.unique(train_data_1200['주로상태_함수율'])
tot_mean=np.mean(train_data_1200['adj_record'])

road0_mean=np.mean(train_data_1200[train_data_1200['주로상태_함수율']==road[0]]['adj_record'])
road1_mean=np.mean(train_data_1200[train_data_1200['주로상태_함수율']==road[1]]['adj_record'])
road2_mean=np.mean(train_data_1200[train_data_1200['주로상태_함수율']==road[2]]['adj_record'])
road3_mean=np.mean(train_data_1200[train_data_1200['주로상태_함수율']==road[3]]['adj_record'])

train_data_1200.loc[train_data_1200['주로상태_함수율']==road[0],'adj_record']=train_data_1200.loc[train_data_1200['주로상태_함수율']==road[0],'adj_record']*tot_mean/road0_mean
train_data_1200.loc[train_data_1200['주로상태_함수율']==road[1],'adj_record']=train_data_1200.loc[train_data_1200['주로상태_함수율']==road[1],'adj_record']*tot_mean/road1_mean
train_data_1200.loc[train_data_1200['주로상태_함수율']==road[2],'adj_record']=train_data_1200.loc[train_data_1200['주로상태_함수율']==road[2],'adj_record']*tot_mean/road2_mean
train_data_1200.loc[train_data_1200['주로상태_함수율']==road[3],'adj_record']=train_data_1200.loc[train_data_1200['주로상태_함수율']==road[3],'adj_record']*tot_mean/road3_mean

test_data_1200.loc[test_data_1200['주로상태_함수율']==road[0],'adj_record']=test_data_1200.loc[test_data_1200['주로상태_함수율']==road[0],'adj_record']*tot_mean/road0_mean
test_data_1200.loc[test_data_1200['주로상태_함수율']==road[1],'adj_record']=test_data_1200.loc[test_data_1200['주로상태_함수율']==road[1],'adj_record']*tot_mean/road1_mean
test_data_1200.loc[test_data_1200['주로상태_함수율']==road[2],'adj_record']=test_data_1200.loc[test_data_1200['주로상태_함수율']==road[2],'adj_record']*tot_mean/road2_mean
test_data_1200.loc[test_data_1200['주로상태_함수율']==road[3],'adj_record']=test_data_1200.loc[test_data_1200['주로상태_함수율']==road[3],'adj_record']*tot_mean/road3_mean
```

```python
tot_mean=np.mean(train_data_1200['adj_record'])
sep1=np.mean(train_data_1200[train_data_1200['습도(%)']<60]['adj_record'])
sep2=np.mean(train_data_1200[(train_data_1200['습도(%)']>=60)&(train_data_1200['습도(%)']<80)]['adj_record'])
sep3=np.mean(train_data_1200[train_data_1200['습도(%)']>=80]['adj_record'])


train_data_1200.loc[train_data_1200['습도(%)']<60,'adj_record']=train_data_1200.loc[train_data_1200['습도(%)']<60,'adj_record']*tot_mean/sep1
train_data_1200.loc[(train_data_1200['습도(%)']>=60)&(train_data_1200['습도(%)']<80),'adj_record']=train_data_1200.loc[(train_data_1200['습도(%)']>=60)&(train_data_1200['습도(%)']<80),'adj_record']*tot_mean/sep2
train_data_1200.loc[train_data_1200['습도(%)']>=80,'adj_record']=train_data_1200.loc[train_data_1200['습도(%)']>=80,'adj_record']*tot_mean/sep3

test_data_1200.loc[test_data_1200['습도(%)']<60,'adj_record']=test_data_1200.loc[test_data_1200['습도(%)']<60,'adj_record']*tot_mean/sep1
test_data_1200.loc[(test_data_1200['습도(%)']>=60)&(test_data_1200['습도(%)']<80),'adj_record']=test_data_1200.loc[(test_data_1200['습도(%)']>=60)&(test_data_1200['습도(%)']<80),'adj_record']*tot_mean/sep2
test_data_1200.loc[test_data_1200['습도(%)']>=80,'adj_record']=test_data_1200.loc[test_data_1200['습도(%)']>=80,'adj_record']*tot_mean/sep3
```
