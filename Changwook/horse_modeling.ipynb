```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('../Jaehoon/fifinal_merge_data.xlsx')
data
```

```python
drop_data=data.drop(columns=['Unnamed: 0','제주4코너통과기록','조교사경력(년)','기수최근1년간1착승률','기수최근1년간2착승률','조교사최근1년간1착승률','조교사최근1년간2착승률','조교사위탁경주마1위승률','조교사위탁경주마2위승률','조교사위탁경주마출전횟수','조교사최근1년간출전횟수'])
drop_data['기수통산승률']=drop_data['기수통산1착승률']+drop_data['기수통산2착승률']
drop_data['조교사통산승률']=drop_data['조교사통산1착승률']+drop_data['조교사통산2착승률']
drop_data=drop_data.drop(columns=['기수통산1착승률','기수통산2착승률','조교사통산1착승률','조교사통산2착승률'])
drop_data=drop_data.drop(columns=['3시간신적설(cm)','mom_total_in3','dad_total_in3','mom_total','dad_total','dad_post','mom_post','mom_rating','dad_rating'])
drop_data=drop_data.drop(columns=['경주거리'])
```

```python
unique = data[["경주일자",	"경주번호"]].drop_duplicates()
unique = unique.reset_index()
unique
```

```python
drop_data["key"]=0
for i in range(len(unique)):
    drop_data.loc[(drop_data["경주일자"]==unique["경주일자"][i])&(drop_data["경주번호"]==unique["경주번호"][i]), "key"] = i
```

```python
drop_data
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score
X_train = drop_data[data["경주일자"]<20220701].drop(["경주기록(초)","경주일자", "착순"], axis=1)
X_test = drop_data[data["경주일자"]>=20220701].drop(["경주기록(초)","경주일자", "착순"], axis=1) 
y_train = drop_data[data["경주일자"]<20220701]["경주기록(초)"]
y_test = drop_data[data["경주일자"]>=20220701]["경주기록(초)"]

params = {
    'max_depth':5,
    'reg_lambda':1,
    'learning_rate': 0.1,
    'n_estimators': 100
}
xgb_reg = XGBRegressor(params = params)
xgb_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = xgb_reg.predict(X_test)
```

```python
pred_key = pd.concat([drop_data[data["경주일자"]>=20220701]["key"], pd.DataFrame(y_pred)], axis=1)
pred_key.columns=["key",	"y_pred"]

true_key = pd.concat([drop_data[data["경주일자"]>=20220701]["key"], pd.DataFrame(y_test)], axis=1)
true_key.columns=["key",	"y_test"]
```

```python
pred_key["in5"] = 0
for i in pred_key["key"].unique():
    index = pred_key[pred_key["key"]==i].sort_values(by="y_pred", ascending=True)[:5].index
    pred_key.loc[index,"in5"]=1
```

```python
true_key["in3"] = 0
for i in true_key["key"].unique():
    index = true_key[true_key["key"]==i].sort_values(by="y_test", ascending=True)[:3].index
    true_key.loc[index,"in3"]=1
```

```python
count = 0
for key in true_key["key"].unique():
    num = sum(true_key.loc[true_key["key"]==key,"in3"]&pred_key.loc[pred_key["key"]==key,"in5"])
    if num==3:
        count+=1
```
