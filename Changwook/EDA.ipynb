### 1. 결측치 처리

먼저 결측치가 있는지 확인해보자. 

```python
for col in data.columns:
    if data[col].isnull().sum()!=0:
        print(col,':',data[col].isnull().sum())
```

7개의 칼럼에서 결측치가 확인되었다. TR_AGE를 제외하고는 다른 모든 칼럼에서 결측치의 개수가 같다. 또 이 칼럼들은 모두 조교사에 관련된 칼럼이다. 조금 더 살펴보자.

```python
data[data["TR_NAME"]=="외부마"]
```

TR_NAME이 외부마로 지정되어 있을 경우 조교사에 관련된 6개의 칼럼이 NaN으로지정되는 것을 확인할 수 있다. 

### 2. 단일 값으로 이루어진 칼럼 처리

모든 행에 대해서 단일 값으로 이루어진 열들이 많이 보인다. 이들은 정보가 없다고 봐도 무방함으로 drop한다. 

```python
unicol_list = []
for col in data.columns:
    if data[1:][col].nunique()<=1:
        unicol_list.append(col)
```

```python
drop_data = data.drop(unicol_list,axis=1) 
drop_data
```

### 3. 기타 칼럼 수정

```python
drop_data["WG"] = drop_data["WG_HR"].str.extract(r'(\d+)(?=\()')
drop_data["WG"][0]="마체중"

drop_data["WC"] = drop_data["WG_HR"].str.extract(r'\((.*?)\)')
drop_data["WC"][0] = "체중증감"

drop_data["WC"][1:] = pd.to_numeric(drop_data["WC"][1:])
drop_data["WG"][1:] = pd.to_numeric(drop_data["WG"][1:])

drop_data = drop_data.drop(["WG_HR"],axis=1)
```
