## Crawling2 만들기

```python
data = pd.read_csv("../Data/crawling.csv")
```

```python
data = data.drop(["Unnamed: 0"], axis=1)
```

```python
data
```

```python
# ini price
ini_price = data["ini_price"]
ini_price.unique()
```

```python
data["ini_price"] = ini_price.str.extract(r'(\d+,\d+)천원')[0].str.replace(',','')
data["ini_price"][-data["ini_price"].isnull()].astype("int")
```

```python
data
```

```python
# sicks
data["sicks"]
```

```python
for i in range(len(data)):
    data["sicks"][i] = ast.literal_eval(data["sicks"][i])
```

```python
data["sicks"][0]
```

```python
for i in range(len(data)):
    my_dict = {}
    # Iterate through the list and split each element by space
    for item in data["sicks"][i]:
        parts = item.split(' ')
        date = parts[-1]  # The last part is the date
        value = ' '.join(parts[:-1])  # Join the rest of the parts
        my_dict[date] = value
    data["sicks"][i]=my_dict
```

```python
data["sicks"]
```

```python
# ranks

data["ranks"][0]
```

```python
for i in range(len(data)):
    data["ranks"][i] = ast.literal_eval(data["ranks"][i])
```

```python
data["ranks"]
```

```python
for i in range(len(data)):
    my_dict = {}
    # Iterate through the list and split each element by space
    for item in data["ranks"][i]:
        parts = item.split(' ')
        value = parts[0]
        date = ' '.join(parts[1:])  # Join the rest of the parts
        my_dict[date] = value
    data["ranks"][i] = my_dict
```
